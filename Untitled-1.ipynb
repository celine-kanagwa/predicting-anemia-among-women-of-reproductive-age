```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import (
    accuracy_score, classification_report, roc_auc_score, roc_curve, auc, confusion_matrix
)
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
import seaborn as sns

def compare_models(X_train_resampled, y_train_resampled, X_test, y_test):
    models = {
        'Random Forest': RandomForestClassifier( random_state=42),
        'Logistic Regression': LogisticRegression( max_iter=1000, random_state=42),
        'SVM': SVC(class_weight='balanced', probability=True, random_state=42),
        'KNN': KNeighborsClassifier(),
        'XGBoost': XGBClassifier(
            scale_pos_weight=10,
            eval_metric='logloss',
            use_label_encoder=False,
            random_state=42
        )
    }

    results = []
    combined_fprs = []
    combined_tprs = []
    combined_labels = []

    def get_subset(X, idx):
        if hasattr(X, "iloc"):
            return X.iloc[idx]
        else:
            return X[idx]

    for name, model in models.items():
        print(f"\nTraining and evaluating {name}...")
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Get prediction scores for ROC
        if hasattr(model, "predict_proba"):
            y_scores = model.predict_proba(X_test)[:, 1]
        else:
            y_scores = model.decision_function(X_test)

        acc = accuracy_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_scores)
        report_dict = classification_report(y_test, y_pred, output_dict=True, zero_division=0)


        # Confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(5, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f'Confusion Matrix - {name}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()

        # Cross-validation ROC with StratifiedKFold = 5
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        tprs = []
        base_fpr = np.linspace(0, 1, 101)

        plt.figure(figsize=(8, 6))
        for train_idx, val_idx in cv.split(X_train, y_train):
            model_cv = model.__class__(**model.get_params())
            model_cv.fit(get_subset(X_train, train_idx), get_subset(y_train, train_idx))

            if hasattr(model_cv, "predict_proba"):
                y_val_scores = model_cv.predict_proba(get_subset(X_train, val_idx))[:, 1]
            else:
                y_val_scores = model_cv.decision_function(get_subset(X_train, val_idx))

            fpr, tpr, _ = roc_curve(get_subset(y_train, val_idx), y_val_scores)
            tpr_interp = np.interp(base_fpr, fpr, tpr)
            tpr_interp[0] = 0.0
            tprs.append(tpr_interp)
            plt.plot(fpr, tpr, lw=1, alpha=0.3, label='Fold ROC curve')

        mean_tpr = np.mean(tprs, axis=0)
        mean_tpr[-1] = 1.0
        mean_auc = auc(base_fpr, mean_tpr)
        plt.plot(base_fpr, mean_tpr, color='b',
                 label=f'Mean ROC (AUC = {mean_auc:.3f})', lw=2)
        plt.plot([0, 1], [0, 1], linestyle='--', color='grey', lw=2)
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'Cross-Validated ROC Curve - {name}')
        plt.legend(loc='lower right')
        plt.show()

        # Save for combined ROC plot
        fpr, tpr, _ = roc_curve(y_test, y_scores)
        combined_fprs.append(fpr)
        combined_tprs.append(tpr)
        combined_labels.append(f'{name} (AUC = {roc_auc:.3f})')

        results.append({
            'Model': name,
            'Accuracy': acc,
            'ROC AUC': roc_auc,
            'Precision (avg)': report_dict['weighted avg']['precision'],
            'Recall (avg)': report_dict['weighted avg']['recall'],
            'F1-score (avg)': report_dict['weighted avg']['f1-score']
        })

    # Combined ROC curve plot
    plt.figure(figsize=(10, 8))
    for fpr, tpr, label in zip(combined_fprs, combined_tprs, combined_labels):
        plt.plot(fpr, tpr, lw=2, label=label)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0, 1])
    plt.ylim([0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=14)
    plt.ylabel('True Positive Rate', fontsize=14)
    plt.title('ROC Curve Comparison (Test Set)', fontsize=16)
    plt.legend(loc='lower right', fontsize=12)
    plt.show()

    df_results = pd.DataFrame(results)
    return df_results
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    roc_auc_score,
    roc_curve,
    auc,
    confusion_matrix
)
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
import lightgbm as lgb


def compare_models_tuned(X_train_resampled, y_train_resampled):
    """
    Compare tuned models with hyperparameter search, plot ROC curves on TRAINING set,
    and return metrics as a DataFrame.
    """
    
    # -------------------- BASE MODELS --------------------
    models = {
        'Random Forest': RandomForestClassifier(random_state=42),
        #'Logistic Regression': LogisticRegression(max_iter=2000, random_state=42),
        'KNN': KNeighborsClassifier(),
        'XGBoost': XGBClassifier(scale_pos_weight=10, eval_metric='logloss', use_label_encoder=False, random_state=42),
        'LightGBM': lgb.LGBMClassifier(class_weight='balanced', random_state=42)
    }

    # -------------------- PARAMETER GRIDS --------------------
    param_grids = {
        'Random Forest': {
            'n_estimators': [100, 200, 300],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5],
            'min_samples_leaf': [1, 2],
            'max_features': ['sqrt', 'log2']
        },
        #'Logistic Regression': {
         #   'C': np.logspace(-2, 2, 10),
          #  'penalty': ['l1'],
           # 'solver': ['liblinear', 'saga'],
            #'class_weight': ['balanced']
        #},
        'KNN': {
            'n_neighbors': list(range(3, 15, 2)),
            'weights': ['uniform', 'distance']
        },
        'XGBoost': {
            'n_estimators': [300, 500],
            'max_depth': [5, 10],
            'learning_rate': [0.05, 0.1],
            'subsample': [0.8, 1.0],
            'colsample_bytree': [0.8, 1.0]
        },
        'LightGBM': {
            'n_estimators': [300, 500],
            'max_depth': [5, 10],
            'learning_rate': [0.05, 0.1],
            'num_leaves': [31, 50],
            'subsample': [0.8, 1.0],
            'colsample_bytree': [0.8, 1.0]
        }
    }

    results = []
    combined_fprs, combined_tprs, combined_labels = [], [], []
    base_fpr = np.linspace(0, 1, 101)

    def get_subset(X, idx):
        return X.iloc[idx] if hasattr(X, "iloc") else X[idx]

    for name, model in models.items():
        print(f"\nTuning {name}...")

        # -------------------- RANDOMIZED SEARCH --------------------
        search = RandomizedSearchCV(
            estimator=model,
            param_distributions=param_grids[name],
            n_iter=15,
            scoring='roc_auc',
            cv=5,
            verbose=0,
            random_state=42,
            n_jobs=-1
        )
        search.fit(X_train_resampled, y_train_resampled)
        best_model = search.best_estimator_
        print(f"Best params: {search.best_params_}")

        # -------------------- TRAINING SET METRICS --------------------
        best_model.fit(X_train_resampled, y_train_resampled)
        y_train_pred = best_model.predict(X_train_resampled)
        if hasattr(best_model, "predict_proba"):
            y_train_scores = best_model.predict_proba(X_train_resampled)[:, 1]
        else:
            y_train_scores = best_model.decision_function(X_train_resampled)

        acc = accuracy_score(y_train_resampled, y_train_pred)
        roc_auc = roc_auc_score(y_train_resampled, y_train_scores)
        report_dict = classification_report(
            y_train_resampled, y_train_pred, output_dict=True, zero_division=0
        )

        # -------------------- CONFUSION MATRIX --------------------
        cm = confusion_matrix(y_train_resampled, y_train_pred)
        plt.figure(figsize=(5, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f'Confusion Matrix - {name} (Training)')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()

        # -------------------- CROSS-VALIDATED ROC (TRAINING) --------------------
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        tprs = []
        plt.figure(figsize=(8, 6))

        for train_idx, val_idx in cv.split(X_train_resampled, y_train_resampled):
            model_cv = best_model.__class__(**best_model.get_params())
            model_cv.fit(get_subset(X_train_resampled, train_idx), get_subset(y_train_resampled, train_idx))

            if hasattr(model_cv, "predict_proba"):
                y_val_scores = model_cv.predict_proba(get_subset(X_train_resampled, val_idx))[:, 1]
            else:
                y_val_scores = model_cv.decision_function(get_subset(X_train_resampled, val_idx))

            fpr, tpr, _ = roc_curve(get_subset(y_train_resampled, val_idx), y_val_scores)
            tpr_interp = np.interp(base_fpr, fpr, tpr)
            tpr_interp[0] = 0.0
            tprs.append(tpr_interp)
            plt.plot(fpr, tpr, lw=1, alpha=0.3)

        mean_tpr = np.mean(tprs, axis=0)
        mean_tpr[-1] = 1.0
        mean_auc = auc(base_fpr, mean_tpr)
        plt.plot(base_fpr, mean_tpr, color='b', label=f'Mean ROC (AUC = {mean_auc:.3f})', lw=2)
        plt.plot([0, 1], [0, 1], linestyle='--', color='grey', lw=2)
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'Cross-Validated ROC Curve - {name} (Training)')
        plt.legend(loc='lower right')
        plt.show()

        # -------------------- COMBINED ROC (TRAINING) --------------------
        fpr, tpr, _ = roc_curve(y_train_resampled, y_train_scores)
        combined_fprs.append(fpr)
        combined_tprs.append(tpr)
        combined_labels.append(f'{name} (AUC = {roc_auc:.3f})')

        # -------------------- STORE RESULTS --------------------
        results.append({
            'Model': name,
            'Accuracy': acc,
            'ROC AUC': roc_auc,
            'Precision (avg)': report_dict['weighted avg']['precision'],
            'Recall (avg)': report_dict['weighted avg']['recall'],
            'F1-score (avg)': report_dict['weighted avg']['f1-score']
        })

    # -------------------- PLOT ALL ROC CURVES TOGETHER --------------------
    plt.figure(figsize=(10, 8))
    for fpr, tpr, label in zip(combined_fprs, combined_tprs, combined_labels):
        plt.plot(fpr, tpr, lw=2, label=label)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate', fontsize=14)
    plt.ylabel('True Positive Rate', fontsize=14)
    plt.title('ROC Curve Comparison (Training Set)', fontsize=16)
    plt.legend(loc='lower right', fontsize=12)
    plt.show()

    return pd.DataFrame(results), best_model
```

```python
from sklearn.feature_selection import mutual_info_classif, chi2
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
import pandas as pd
import numpy as np

def feature_selection_table(X_train, y_train, feature_names, top_n=20):
    X_df = pd.DataFrame(X_train, columns=feature_names)
    y_series = pd.Series(y_train, index=X_df.index)

    # Drop NaN targets
    valid_idx = y_series.notna()
    X_df = X_df.loc[valid_idx]
    y_series = y_series.loc[valid_idx]

    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X_df)

    # 1) Mutual Info scores
    mi_scores = mutual_info_classif(X_df, y_series, random_state=42)
    mi_top_features = list(X_df.columns[np.argsort(mi_scores)[::-1][:top_n]])

    # 2) Chi2 scores
    chi2_scores, _ = chi2(X_scaled, y_series)
    chi2_top_features = list(X_df.columns[np.argsort(chi2_scores)[::-1][:top_n]])

    # 3) Logistic Regression Coefs
    model_log = LogisticRegression(max_iter=1000, solver='liblinear')
    model_log.fit(X_df, y_series)
    coefs = np.abs(model_log.coef_[0])
    coef_top_features = list(X_df.columns[np.argsort(coefs)[::-1][:top_n]])

    # 4) Stepwise forward selection using mlxtend
    model_fs = LogisticRegression(max_iter=1000)
    sfs_forward = SFS(model_fs,
                      k_features=top_n,
                      forward=True,
                      floating=False,
                      scoring='accuracy',
                      cv=3,
                      n_jobs=-1,
                      verbose=0)
    sfs_forward = sfs_forward.fit(X_df.values, y_series.values)
    step_forward_feats = [feature_names[i] for i in sfs_forward.k_feature_idx_]

    # 5) Stepwise backward selection using mlxtend
    sfs_backward = SFS(model_fs,
                       k_features=top_n,
                       forward=False,
                       floating=False,
                       scoring='accuracy',
                       cv=3,
                       n_jobs=-1,
                       verbose=0)
    sfs_backward = sfs_backward.fit(X_df.values, y_series.values)
    step_backward_feats = [feature_names[i] for i in sfs_backward.k_feature_idx_]

    # Ensure each column has top_n rows, pad with ''
    def pad_list(l):
        return l + [''] * (top_n - len(l))

    mi_col = pad_list(mi_top_features)
    chi2_col = pad_list(chi2_top_features)
    coef_col = pad_list(coef_top_features)
    step_f_col = pad_list(step_forward_feats)
    step_b_col = pad_list(step_backward_feats)


    from sklearn.model_selection import cross_val_score
    def rf_accuracy(selected_feats, cv=5):
       if not selected_feats:
        return np.nan
       rf = RandomForestClassifier(random_state=42)
       scores = cross_val_score(rf, X_df[selected_feats], y_series, cv=cv, scoring='accuracy')
       return scores.mean()
    
    acc_mi = rf_accuracy(mi_top_features)
    acc_chi2 = rf_accuracy(chi2_top_features)
    acc_coef = rf_accuracy(coef_top_features)
    acc_step_f = rf_accuracy(step_forward_feats)
    acc_step_b = rf_accuracy(step_backward_feats)

    # Create dataframe
    df = pd.DataFrame({
        'Mutual Info': mi_col,
        'Chi2': chi2_col,
        'Classif': coef_col,
        'Step Forward': step_f_col,
        'Step Backward': step_b_col
    })

    # Add accuracy row
    accuracy_row = pd.DataFrame({
        'Mutual Info': [f"Accuracy: {acc_mi:.3f}"],
        'Chi2': [f"Accuracy: {acc_chi2:.3f}"],
        'Classif': [f"Accuracy: {acc_coef:.3f}"],
        'Step Forward': [f"Accuracy: {acc_step_f:.3f}"],
        'Step Backward': [f"Accuracy: {acc_step_b:.3f}"]
    })

    df = pd.concat([df, accuracy_row], ignore_index=True)
    return df
```
